# Практикум 1. Подготовка рабочей среды

Прежде чем переходить к анализу энергетических данных, нужно убедиться, что рабочий контур собран корректно: репозиторий открыт из нужной папки, библиотеки доступны, а относительные пути действительно ведут к данным и материалам пособия.

## Зачем выполняется эта работа

В учебных и инженерных блокнотах ошибки часто возникают не из-за методов анализа, а из-за плохо подготовленной среды. Если корень проекта определен неверно или ядро зарегистрировано не в том окружении, следующие разделы начинают давать несогласованные результаты.

## Что будет сделано в практикуме

- будет определен фактический корень учебного репозитория;
- будут зафиксированы версии ключевых библиотек;
- будет проверено наличие основных каталогов курса;
- будет сделан вывод о готовности среды к дальнейшей работе.

## Данные и инструменты

| Компонент | Роль в практикуме |
| --- | --- |
| `Path` | определение корня проекта и проверка структуры каталогов |
| `pandas` | вывод компактных диагностических таблиц |
| `display_utils` | аккуратное представление результатов |
| учебный репозиторий | единая точка доступа к блокнотам, данным и пособию |

## Ход работы

### Шаг 1. Установить, из какого каталога фактически открыт блокнот
Если этот шаг пропустить, далее будут нарушены относительные пути к данным и рисункам.

### Шаг 2. Зафиксировать вычислительное окружение
Версии Python и библиотек нужно записывать так же аккуратно, как и результаты анализа.

### Шаг 3. Проверить структуру учебного комплекта
Только после этого можно уверенно переходить к данным, визуализации и моделям.

```python
from pathlib import Path
import platform
import sys

import numpy as np
import pandas as pd

from src.display_utils import display_callout, display_frame, display_metric_table, display_stage_note

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
display_stage_note(
    "Шаг 1. Корень проекта",
    "Сначала зафиксируем путь, из которого блокнот обращается к данным и остальным материалам курса.",
)

project_location = pd.DataFrame(
    [{"Текущий каталог": str(Path.cwd()), "Корень проекта": str(PROJECT_ROOT)}]
)
display_frame(project_location, decimals=0)
```

## Пояснение к первому шагу

Если блокнот открыт из каталога `notebooks`, вычисляется переход на уровень выше. Такая схема позволяет использовать единые относительные пути во всем репозитории и не переписывать код при переносе материалов на другую машину.

```python
display_stage_note(
    "Шаг 2. Вычислительное окружение",
    "Теперь зафиксируем версии Python и основных библиотек, чтобы последующие результаты были воспроизводимы.",
)

environment_info = pd.DataFrame(
    [
        {
            "Python": sys.version.split()[0],
            "Платформа": platform.platform(),
            "pandas": pd.__version__,
            "NumPy": np.__version__,
        }
    ]
)
display_frame(environment_info, decimals=0)
```

## Как читать таблицу окружения

Эта таблица не является формальностью. При расхождении версий библиотек могут отличаться детали чтения файлов, настройки графиков и даже численные результаты при обучении моделей. Поэтому окружение нужно воспринимать как часть исследовательского протокола.

```python
display_stage_note(
    "Шаг 3. Структура учебного комплекта",
    "Проверим, присутствуют ли каталоги, без которых не будет работать связка пособия, данных, кода и учебного ядра.",
)

project_dirs = [
    PROJECT_ROOT / "docs",
    PROJECT_ROOT / "notebooks",
    PROJECT_ROOT / "data",
    PROJECT_ROOT / "src",
    PROJECT_ROOT / "kernel",
    PROJECT_ROOT / "handbook",
]

status = pd.DataFrame(
    {
        "Каталог": [str(path.relative_to(PROJECT_ROOT)) for path in project_dirs],
        "Найден": [path.exists() for path in project_dirs],
    }
)
display_frame(status, decimals=0)

display_metric_table(
    {
        "каталогов проверено": float(len(project_dirs)),
        "каталогов найдено": float(status["Найден"].sum()),
        "каталогов отсутствует": float((~status["Найден"]).sum()),
    },
    decimals=0,
)
```

## Промежуточный вывод

Если ключевые каталоги найдены, рабочий контур можно считать собранным корректно. Это означает, что в следующих практикумах данные будут загружаться локально, а гиперссылки, графики и сопроводительные материалы останутся согласованными.

```python
missing_catalogs = status.loc[~status["Найден"], "Каталог"].tolist()

if missing_catalogs:
    display_callout(
        "Среда требует донастройки",
        "Не найдены каталоги: " + ", ".join(missing_catalogs) + ". Перед продолжением их нужно восстановить.",
        tone="warning",
    )
else:
    display_callout(
        "Среда готова к работе",
        "Корень проекта определен корректно, а структура учебного комплекта согласована с материалами пособия.",
        tone="success",
    )
```
