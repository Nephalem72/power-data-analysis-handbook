# 🚀 Практикум 1. Подготовка рабочей среды и проверка структуры курса

Перед началом работы с энергетическими данными важно убедиться, что учебный контур собран корректно: окружение установлено, репозиторий открыт из правильной директории, а демонстрационные наборы данных доступны по относительным путям.


## 🧭 Введение и описание практикума

В реальной исследовательской работе ошибки часто появляются не на этапе моделирования, а гораздо раньше: блокнот открыт не из того каталога, пакеты установлены в другом окружении, пути к файлам зашиты вручную и перестают работать после переноса проекта на другой компьютер. Поэтому стартовый практикум нужен как методическая страховка: он делает дальнейшую работу воспроизводимой.


### 🎯 Цель практикума

Установить, что рабочая директория, структура репозитория и базовый набор библиотек согласованы между собой и позволяют без дополнительных ручных правок запускать остальные учебные блокноты.


### ✨ Почему этот этап нельзя пропускать

Если на старте не проверить среду, то последующие ошибки будут маскироваться под «проблемы модели» или «ошибки данных», хотя фактическая причина окажется чисто инфраструктурной. Для учебного курса это особенно критично: читатель должен видеть устойчивый, повторяемый результат, а не бороться с несогласованностью путей и зависимостей.


### 🧩 Задачи практикума

1. Зафиксировать корень проекта и проверить учебные каталоги.
2. Убедиться, что основные библиотеки доступны в текущем окружении.
3. Проверить наличие демонстрационных наборов данных.
4. Подготовить опорную информацию для дальнейших практикумов.

### 🔎 Как читать стартовую сводку

Из этой таблицы должно быть видно, что работа ведется внутри репозитория курса, а не в произвольной временной папке. Если корень проекта определен неверно, то уже на следующем шаге блокнот перестанет находить наборы данных и вспомогательные модули.


```python
from pathlib import Path
import sys

import pandas as pd

# Добавляем корень репозитория в путь импорта, чтобы ноутбук запускался
# одинаково и из корня проекта, и из папки notebooks.
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from src.project_paths import project_root, sample_data_dir
from src.display_utils import display_callout, display_frame, display_metric_table, display_stage_note


# Проверяем ключевые каталоги курса и формируем учебный реестр структуры проекта.
root = project_root()
folders = ["notebooks", "data", "docs", "src", "tasks", "handbook"]

structure = pd.DataFrame(
    {
        "Каталог": folders,
        "Найден": [(root / folder).exists() for folder in folders],
        "Путь": [str((root / folder).resolve()) for folder in folders],
    }
)

display_stage_note(
    "Структура репозитория",
    "На этом этапе проверяется, что учебные материалы, исходные данные и текст пособия лежат в ожидаемых каталогах.",
)
display_frame(structure)
```

### 🗃 Почему структура каталога так важна

Линейная и прозрачная структура репозитория делает курс удобным для клонирования и воспроизведения. Читатель должен понимать, где искать блокноты, где лежат данные, куда помещены исходники пособия и в каком каталоге находятся вспомогательные модули. В учебной практике это снижает порог входа и делает каждое действие логически предсказуемым.


```python
from pathlib import Path
import sys

import pandas as pd

# Добавляем корень репозитория в путь импорта, чтобы ноутбук запускался
# одинаково и из корня проекта, и из папки notebooks.
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from src.project_paths import project_root, sample_data_dir
from src.display_utils import display_callout, display_frame, display_metric_table, display_stage_note


import importlib

# Проверяем, что основные библиотеки доступны в выбранном ядре.
packages = ["pandas", "numpy", "matplotlib", "seaborn", "sklearn", "openpyxl"]
package_status = []

for package in packages:
    try:
        module = importlib.import_module(package)
        version = getattr(module, "__version__", "версия не определена")
        package_status.append((package, True, version))
    except Exception:
        package_status.append((package, False, "не найдено"))

package_frame = pd.DataFrame(
    package_status, columns=["Библиотека", "Доступна", "Версия"]
)

display_stage_note(
    "Проверка библиотек",
    "Теперь можно убедиться, что учебное ядро действительно содержит пакеты, которые понадобятся в следующих разделах.",
)
display_frame(package_frame)
```

### 📦 Что показывает таблица библиотек

Если основные библиотеки доступны уже на стартовом практикуме, то далее можно сосредоточиться на методике анализа, а не на исправлении окружения. Особенно важны `openpyxl` для чтения `XLSX`, `seaborn` для аккуратной визуализации и `scikit-learn` для первых моделей и метрик качества.


```python
from pathlib import Path
import sys

import pandas as pd

# Добавляем корень репозитория в путь импорта, чтобы ноутбук запускался
# одинаково и из корня проекта, и из папки notebooks.
ROOT = Path.cwd().resolve().parent if Path.cwd().name == "notebooks" else Path.cwd().resolve()
if str(ROOT) not in sys.path:
    sys.path.append(str(ROOT))

from src.project_paths import project_root, sample_data_dir
from src.display_utils import display_callout, display_frame, display_metric_table, display_stage_note


# Собираем краткий перечень учебных наборов данных, которые будут использоваться далее.
sample_dir = sample_data_dir()
dataset_rows = []

for file_path in sorted(sample_dir.rglob("*")):
    if file_path.is_file() and file_path.name != ".gitkeep":
        dataset_rows.append(
            {
                "Файл": file_path.name,
                "Раздел": file_path.parent.name,
                "Размер, КБ": round(file_path.stat().st_size / 1024, 1),
            }
        )

datasets = pd.DataFrame(dataset_rows)

display_stage_note(
    "Доступность учебных данных",
    "Эта таблица подтверждает, что после обычного `git clone` практикумы увидят локальные копии демонстрационных наборов данных.",
)
display_frame(datasets)
```
